```python
import pandas as pd

# 1. Load your files
exp = pd.read_csv("Occupations_unique_cleaned.csv")      # exposure per occupation
skills = pd.read_csv("Occupation_skills_clean.csv")      # occupations + skills + exposure

# Quick check of columns 
print(exp.columns)
print(skills.columns)

# 2. Merge on O*NET-SOC Code
merged = exp.merge(
    skills[["O*NET-SOC Code", "Skills"]],   # we only need the Skills + code from this file
    on="O*NET-SOC Code",
    how="left"                              # keep all occupations from the exposure file
)

# 3. How many occupations have skills (Skills NOT missing)?
num_with_skills = merged["Skills"].notna().sum()
print("Number of occupations with both exposure score and skills:", num_with_skills)

# 4. How many occupations are missing skills?
num_missing_skills = merged["Skills"].isna().sum()
print("Number of occupations missing skills:", num_missing_skills)
```

```python
# 5. (Optional) See which occupations are missing skills
missing_skills = merged[merged["Skills"].isna()]
print(missing_skills.head())

# 6. (Optional) Save the merged file for later use
merged.to_csv("Merged.csv", index=False)
```

```python
dfm = pd.read_csv("Merged.csv")
dfm.isna().sum()
```

```python
# Drop rows with missing Skills
dfm_cleaned = dfm.dropna(subset=["Skills"])
```
