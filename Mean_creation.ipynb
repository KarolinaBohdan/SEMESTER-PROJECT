```python
import pandas as pd

# Loading merged file
df = pd.read_csv("ILO_Occupations_with_exposure_score.csv")
```

#### Calculating and average exposure score per occupation

```python
# Computing the average exposure per occupation
occupation_mean = (
    df.groupby("O*NET-SOC Code")["Mean"]
      .mean()
      .reset_index()
      .rename(columns={"Mean": "Final_ExposureScore"})
)

# Merging this average back into the full dataframe
df_final = df.merge(occupation_mean, on="O*NET-SOC Code", how="left")

# 4. Save the final file
df_final.to_csv("Occupations_with_FinalExposure.csv", index=False)

df_final.head(10)
```

#### Cleaning

```python

# 1. Load the file you just created
df_final = pd.read_csv("Occupations_with_FinalExposure.csv")

# 2. Drop the columns you don't want
df_cleaned = df_final.drop(columns=[
    "soc2019_code", 
    "soc2010_code",
    "soc2010_6",
    "isco08_code",
    "isco08_title",
    "4-digit code",
    "Exposure",
    "Mean"
])

# 3. Save cleaned file
df_cleaned.to_csv("Occupations_clean.csv", index=False)

df_cleaned.head()
```

```python
import pandas as pd

df1 = pd.read_csv("Occupations_clean.csv")

# Drop duplicates by occupation
df_unique = df1.drop_duplicates()

# Save
df_unique.to_csv("Occupations_unique.csv", index=False)

df_unique.head() 
```
